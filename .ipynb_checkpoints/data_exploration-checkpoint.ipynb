**Stats by position?**

```python
all_df.groupby('Pos').mean()
```

Comments:
- Point Guards most PTS, AST, TO, STL, Drives, USG %
- Shooting Guards best at shooting 3PT
- Small Forward do not excel in any area, balanced (more similar to Guards)
- Power Forards do not excel in any area, balanced (more similar to Centers)
- Centers more REB, BLK

- PTS among different positions are relatively balanced (between 12.71 and 14.26)

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
pd.set_option('display.max_columns',30)
```

```python
df_adv = pd.read_csv('data/advanced/2020_advanced.csv')
df_cas = pd.read_csv('data/catch_and_shoot/2020_catchshoot.csv')
df_driv = pd.read_csv('data/drives/2020_drives.csv')
df_per100 = pd.read_csv('data/per100/2020_per100.csv')
df_post = pd.read_csv('data/postups/2020_postups.csv')
df_shot = pd.read_csv('data/shot_location/2020_shooting.csv')
df_trad = pd.read_csv('data/traditional/2020_traditional.csv')
```

```python
df_trad = df_trad.loc[df_trad['MIN'] > 20]
```

```python
df = df_trad.merge(df_shot, how='inner', left_on='Player', right_on='Player')
```

```python
df = df.merge(df_post, how='inner', left_on='Player', right_on='Player')
```

```python
df = df.merge(df_driv, how='inner', left_on='Player', right_on='Player')
```

```python
df = df.merge(df_cas, how='inner', left_on='Player', right_on='Player')
```

```python
df = df.merge(df_per100, how='inner', left_on='Player', right_on='Player')
```

```python
df = df.merge(df_adv, how='inner', left_on='Player', right_on='Player')
```

```python
df.replace('-', '0', inplace=True)
```

```python
temp = df.loc[:, df.columns != 'Player'].astype('float')
```

```python
df = pd.concat([df['Player'], temp], axis=1)
```

```python
df
```

```python
df = df.drop('Minutes', axis=1)
```

```python
df.to_csv('data/data_without_heights/2020_df', index=False)
```

```python
df_final = pd.read_csv('data/data_without_heights/2020_df')
df_heights = pd.read_csv('data/heights/2020_heights.csv')
```

```python
df_heights['Height'] = [df_heights['Height'][i].split('-') for i in range(len(df_heights))]
```

```python
df_heights['Height'] = [(int(df_heights['Height'][i][0]) * 12) + 
                        int(df_heights['Height'][i][1]) for i in range(len(df_heights))]
```

```python
df_final = df_final.merge(df_heights, how='inner', left_on='Player', right_on='Player')
```

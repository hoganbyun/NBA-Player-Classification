```python
# add year label to player name of each DataFrame
# combine year-by-year data into singular DataFrame

df_2015 = pd.read_csv('data/finalized_data/2015_df')
df_2016 = pd.read_csv('data/finalized_data/2016_df')
df_2017 = pd.read_csv('data/finalized_data/2017_df')
df_2018 = pd.read_csv('data/finalized_data/2018_df')
df_2019 = pd.read_csv('data/finalized_data/2019_df')
df_2020 = pd.read_csv('data/finalized_data/2020_df')
```

```python
df_2015['Player'] = df_2015['Player'] + ': 2015'
df_2016['Player'] = df_2016['Player'] + ': 2016'
df_2017['Player'] = df_2017['Player'] + ': 2017'
df_2018['Player'] = df_2018['Player'] + ': 2018'
df_2019['Player'] = df_2019['Player'] + ': 2019'
df_2020['Player'] = df_2020['Player'] + ': 2020'
```

```python
df_2020.head()
```

```python
# before we concat everything, drop players who have played less than 20% of a team's games
# 16 games for 2015-2018 (82 games), 14 games for 2019 (league average 71 games), 8 games for 2020 (LA 42 games)
df_2015.drop(df_2015[df_2015['GP'] < 16].index, inplace=True)
df_2016.drop(df_2016[df_2016['GP'] < 16].index, inplace=True)
df_2017.drop(df_2017[df_2017['GP'] < 16].index, inplace=True)
df_2018.drop(df_2018[df_2018['GP'] < 16].index, inplace=True)
df_2019.drop(df_2019[df_2019['GP'] < 14].index, inplace=True)
df_2020.drop(df_2020[df_2020['GP'] < 8].index, inplace=True)
```

```python
all_df = pd.concat([df_2015, df_2016, df_2017, df_2018, df_2019, df_2020])
```
